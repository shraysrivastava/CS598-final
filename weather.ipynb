```python
import requests
import datetime
from datetime import datetime
```

```python
def get_weather_data(lat, lon, date_time, timezone='America/New_York'):
    base_url = "https://archive-api.open-meteo.com/v1/archive"
    date_str = date_time.strftime('%Y-%m-%d')
    params = {
        'latitude': lat,
        'longitude': lon,
        'start_date': date_str,
        'end_date': date_str,
        'hourly': 'temperature_2m,precipitation,weathercode',
        'timezone': timezone
    }
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        data = response.json()
        target_time = date_time.strftime('%Y-%m-%dT%H:%M')
        times = data['hourly']['time']
        if target_time in times:
            index = times.index(target_time)
            weather_info = {
                'temperature': data['hourly']['temperature_2m'][index],
                'precipitation': data['hourly']['precipitation'][index],
                'weathercode': data['hourly']['weathercode'][index]
            }
            return weather_info
        else:
            print(f"Time {target_time} not found in API response.")
            return None
    except requests.exceptions.RequestException as e:
        print(f"API request error: {e}")
        return None
    except KeyError as e:
        print(f"Key error: {e}")
        return None
```

```python
def map_weathercode_to_description(weathercode):
    if weathercode in [0, 1]:
        return 'Clear'
    elif weathercode in [2, 3]:
        return 'Cloudy'
    elif weathercode in [45, 48]:
        return 'Fog/Haze'
    elif weathercode in [51, 53, 55, 56, 57]:
        return 'Drizzle'
    elif weathercode in [61, 80]:
        return 'Light Rain'
    elif weathercode in [63, 65, 66, 67, 81, 82]:
        return 'Heavy Rain'
    elif weathercode in [71, 85]:
        return 'Light Snow'
    elif weathercode in [73, 75, 77, 86]:
        return 'Heavy Snow'
    elif weathercode in [95, 96, 99]:
        return 'Thunderstorm'
    else:
        return 'Unknown'
```

```python

# Sample data
latitude = 40.7128
longitude = -74.0060
date_time = datetime(2023, 1, 1, 15, 0)


weather_data = get_weather_data(latitude, longitude, date_time)

if weather_data:
    weather_description = map_weathercode_to_description(weather_data['weathercode'])
    print(f"Temperature: {weather_data['temperature']}°C")
    print(f"Precipitation: {weather_data['precipitation']} mm")
    print(f"Weather Condition: {weather_description}")
```
